# Sonifying ApRES data
As described [here](https://ldeo-glaciology.github.io/glaciology-intro-book/sections/radar/apres/theory_1.html#compute-the-frequency-difference), the deramped signal recorded by ApRES (i.e. the chirps) have frequencies in the kHz range, which covers the range of frequencies that are audible to humans. Therefore, chirps can be straightforwardly 'sonified', i.e. turned in to sound. 


```python
import sys
sys.path.append("../../../xapres") 
import xapres as xa
import matplotlib.pyplot as plt
data = xa.load.generate_xarray(directory='../data/thwaites/');
signal_to_sonfiy = data.chirp.isel(time=0, chirp_num=0, attenuator_setting_pair=0)
signal_to_sonfiy.plot(figsize=(20, 5));
plt.title('signal to sonify');
```


The method `.sonify()` is added to xarrays produced by xapres. It plays the chirp through your computer's speakers and, optionally, saves the sound in a common sound format (.wav). 

It requires two additional packages that can be installed with

The following calls the sonify method.

```python
signal_to_sonfiy.sonify(save = True)
```

This should have played a one second sound and produced a new file called 'chirp.wav'
